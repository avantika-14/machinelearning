```python
dataset.columns=['Region','Date','Frequency','Estimated Unemployment Rate (%)','Estimated Employed',
                 'Estimated Labour Participation Rate (%)','Region.1','longitude','latitude'
                  ]
plt.title("Unemployment in India")
sns.histplot(x="Estimated Employed",hue="Region.1",data=dataset)
plt.show()
```

```python
unemployment=dataset[["Region","Region.1","Estimated Unemployment Rate (%)"]]
fig=px.sunburst(unemployment,path=["Region.1","Region"],values="Estimated Unemployment Rate (%)",
                color_continuous_scale="RdVIGn",title='Unemployment Rate in India')
fig.show()
```
